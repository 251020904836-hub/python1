```python
a = float(input("nhập số a:"))
b = float(input("nhập số b:"))
c = float(input("nhập số c:"))
if a == 0 :
  if b == 0:
    print("Phương trình vô nghiệm")
  else :
    x = -c/b
    print("Phương trình có nghiện là:",x)
else:
  delta = b**2 - 4*a*c
  print("Delta:",delta)
  if delta > 0:
    x1 = (-b + delta**0.5)/(2*a)
    x2 = (-b - delta**0.5)/(2*a)
    print("Phương trình có 2 nghiệm phân biệt")
    print("nghiệm 1 =",x1)
    print("nghiệm 2 =",x2)
  elif delta == 0:
    x = -b/(2*a)
    print("Phương trình có 1 nghiệm kép x=",x)
  else :
    print("Phương trình vô nghiệm")
```
